```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
#plt.switch_backend('agg')
import mpl_stylesheet
mpl_stylesheet.banskt_presentation(fontfamily = 'latex', fontsize = 38, linewidth = 4)

colors=['#2A53C1', '#F03C3F']
```

```python
import scipy


figsize=(12,5)
fig = plt.figure(figsize=figsize)
ax = fig.add_subplot(1,1,1)
x = np.linspace(-8, 8, num=200)
y1 = 1 / (1 + np.exp(-x))
y2 = 1 - y1
ax.axhline(y=1, xmin=-8, xmax=8, color='black', ls='--')
ax.plot(x, y1, color=colors[0])


ax.set_ylim([0, 1.2])
ax.set_xlim([-8.2,8])

# set the spines
ax.spines['left'].set_position(('axes', 0))
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')


ax.set_xlabel(r'Effect of all SNPs $\left( \beta_0 + \sum_{i} \beta_i x_{ni} \right)$')
mlabel1 = r'$p(\,y_{n} = 1 | \mathbf{x}_{n}, \boldsymbol{\beta})$'

xpos = 1.2
ypos1 = 1 / (1 + np.exp(-xpos)) - 0.25
ax.text(xpos, ypos1, mlabel1, color=colors[0])

ax.set_xticks([])
ax.set_yticks([0, 1])

    
plt.tight_layout()

plt.savefig('../images/logistic_function.pdf', bbox_inches='tight')
plt.show()
```

```python
colors=['#2A53C1', '#F03C3F', '#00B4FF']
axcol='#A3A3A3'

figsize = (12,8)
fig = plt.figure(figsize=figsize)
ax1 = fig.add_subplot(1,1,1)


x = np.linspace(-6, 6, num=2000)
mu = [0.0, 0.0, 1.2]
sigma = [0.1, 1.8, 0.8]
y = [matplotlib.mlab.normpdf(x, mu[i], sigma[i]) for i in range(3)]

ax1.plot(x, 0.5 * y[2], color=colors[2])
ax1.fill(x, 0.5 * y[2], color=colors[2], alpha=0.2)
ax1.plot(x, 0.3 * y[0], color=colors[0])
ax1.plot(x, y[1], color=colors[1])

mlabel1 = r'Prior probability''\n'r'for non-causal SNPs''\n'r'$\mathcal{N}\left( \beta_{i} | 0, \sigma_{\mathrm{bg}}^{2} \right)$'
mlabel2 = r'Prior probability''\n'r'for causal SNPs''\n'r'$\mathcal{N}\left( \beta_{i} | 0, \sigma_{i}^{2} \right)$'
mlabel3 = r'Likelihood''\n'r'$\mathcal{L}(\beta_{0}, \beta_{i})$'
ax1.text(0.5, 0.7, mlabel1, {'color':colors[0]})
ax1.text(-6.1, 0.22, mlabel2, {'color':colors[1]})
ax1.text(2.0, 0.2, mlabel3, {'color':colors[2]})

ax1.set_ylim([0, 1.5])
ax1.set_xlim([-6, 6])

# set the splines
ax1.spines['left'].set_position('zero')
ax1.spines['bottom'].set_position('zero')
ax1.spines['right'].set_color('none')
# ax1.yaxis.tick_left()
ax1.spines['top'].set_color('none')
#ax1.xaxis.tick_bottom()


ax1.set_xlabel(r'Effect size $(\beta_{i})$')



#addticks(ax1, [], [])
ax1.set_yticks([])
ax1.set_xticks([])

#ax1.set_xticklabels([0])



plt.tight_layout()

plt.savefig('../images/bayesian_logistic_regression.pdf', bbox_inches='tight')
plt.show()
```
